```python
import pandas as pd
import pickle

# Load dataset
df = pd.read_csv("preprocessed_heart_data.csv").reset_index(drop=True)

y = df["IsHeartDisease"]
X = df.drop(columns=["IsHeartDisease"])

# Load fixed split (tuple format)
with open("split_indices.pkl", "rb") as f:
    train_idx, val_idx = pickle.load(f)

# Use positional indexing
X_train = X.iloc[train_idx]
X_val   = X.iloc[val_idx]
y_train = y.iloc[train_idx]
y_val   = y.iloc[val_idx]

print("Training shape:", X_train.shape)
print("Validation shape:", X_val.shape)
```

```python
import hashlib

def hash_dataframe(df):
    return hashlib.md5(pd.util.hash_pandas_object(df, index=True).values).hexdigest()

print("Train Hash:", hash_dataframe(X_train))
print("Validation Hash:", hash_dataframe(X_val))
```
